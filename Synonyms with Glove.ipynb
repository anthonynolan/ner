```python
with open('data/glove/glove.6B.50d.txt', 'rt') as f:
    glove = f.readlines()
print(len(glove))
```

```python
import numpy as np
```

```python
small_glove = {}
all_embeddings = []
full_vocab = []
for line in glove:
    word = line.split()[0].lower()
    embedding = np.array(line.split()[1:], dtype=np.float64)
#     if word in vocab:
    small_glove[word] = {'vector': embedding}
    all_embeddings.append(embedding)
    full_vocab.append(word)
```

```python
embs = np.array(all_embeddings)
```

```python
maxes = np.max(embs, axis = 0)
mins = np.min(embs, axis = 0)
```

```python
hyperplanes = []
for a in range(5):
    hyperplane = []
    for a, b in zip(mins, maxes):
        hyperplane.append(np.random.uniform(a, b))
    hyperplanes.append(np.array(hyperplane))
hyperplanes
```

```python
small_glove.get('pigss', {'vector': 0})['vector']
```

```python
# words = ['the', 'pig', 'dog', 'cat', 'fox', 'wolf']
buckets = {}
for word in full_vocab:

    keys = []
    for hyperplane in hyperplanes:
#         products = hyperplane*small_glove.get(word, {'vector': 0})['vector']
#         hashcount = 0
#         for i, product in enumerate(products):
#             if product>=0:
#                 hashcount+= (2**1)*1
        keys.append(np.sign(np.sum(hyperplane*small_glove.get(word, {'vector': 0})['vector'])))
    key = (tuple(keys))
    if key not in buckets.keys():
        buckets[key] = [word]
    else:
        buckets[key].append(word)
    if word in small_glove.keys():
        small_glove[word]['key'] = key
buckets
```

```python
for key, value in buckets.items():
    print(key, len(value))
```

```python
sorted(buckets.keys())
```

```python
from sklearn.metrics.pairwise import cosine_similarity

target_word_vector = small_glove['pig']['vector']

candidates = []
for candidate in buckets[small_glove['pig']['key']]:
    candidates.append((cosine_similarity(small_glove[candidate]['vector'].reshape(1,-1), target_word_vector.reshape(1,-1)), candidate))    
```

```python
from operator import itemgetter
sorted(candidates, key = itemgetter(0), reverse=True)
```
