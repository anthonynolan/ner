```python
import nltk
import numpy as np

from tqdm.notebook import tqdm, trange
import time

with open('data/bleak.txt', 'rt') as f:
    bleak = f.read()
    
words = nltk.word_tokenize(bleak)
words = [word.lower() for word in words]
```

```python
with open('tokenized_text.txt', 'wt') as f:
    f.write(' '.join(words))
```

```python
vocab = set(words)
print(len(words))
print(len(vocab))
```

```python
glove_400k = 'data/glove/glove.6B.50d.txt'
glove_1_9MM = 'data/glove/glove.42B.300d.txt'
with open(glove_1_9MM, 'rt') as f:
    glove = f.readlines()
print(len(glove))
```

```python
small_glove = {}
all_embeddings = []
full_vocab = []

counter = 0
for line in tqdm(glove):
    if not counter%10000:
        print(counter)
    counter+=1
    word = line.split()[0].lower()
    embedding = np.array(line.split()[1:], dtype=np.float64)
    small_glove[word] = {'vector': embedding}
    all_embeddings.append(embedding)
    full_vocab.append(word)
len(small_glove.keys())
# vocab - small_glove.keys()
```

```python
small_glove['dog']
```

```python
embs = np.array(all_embeddings)
maxes = np.max(embs, axis = 0)
mins = np.min(embs, axis = 0)
```

```python
hyperplanes = []
for a in trange(7):
    hyperplane = []
    for a, b in zip(mins, maxes):
        hyperplane.append(np.random.uniform(a, b))
    hyperplanes.append(np.array(hyperplane))
```

```python
buckets = {}
for word in tqdm(full_vocab):
    keys = []
    for hyperplane in hyperplanes:
        keys.append(np.sign(np.sum(hyperplane*small_glove.get(word, {'vector': 0})['vector'])))
    key = (tuple(keys))
    if key not in buckets.keys():
        buckets[key] = [word]
    else:
        buckets[key].append(word)
    if word in small_glove.keys():
        small_glove[word]['key'] = key
buckets
```

```python
for key, value in buckets.items():
    print(key, len(value))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

target_word = 'truculent'
target_word_vector = small_glove[target_word]['vector']

candidates = []
for candidate in tqdm(buckets[small_glove[target_word]['key']]):
    candidates.append((cosine_similarity(small_glove[candidate]['vector'].reshape(1,-1), target_word_vector.reshape(1,-1)), candidate))    
```

```python
from operator import itemgetter
sorted(candidates, key = itemgetter(0), reverse=True)
```
